```python
## This code is only available to scrape with Chrome, and make sure you have downloaded Chrome Driver
import csv
from getpass import getpass
from time import sleep
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
from msedge.selenium_tools import Edge, EdgeOptions
import numpy as np


## References:
#https://stackoverflow.com/questions/46026987/selenium-gives-selenium-common-exceptions-webdriverexception-message-unknown
#https://chromedriver.chromium.org

#"Xinyu69120230"
```

```python
Times_of_scrolling = 10  #set scrolling pages



#create a function to get data of one post, it should be used after we find the node related to tweets
def tweet_info(tweet):
    users = tweet.find_element_by_xpath('.//span').text
    
    ## Now we try to exclude posts with no dates, like ads...
    ## These ads doesn't have a date time
    try:
        Date = tweet.find_element_by_xpath('.//time')  ##It is all about finding the right node...
        Date = Date.get_attribute("datetime")
    except NoSuchElementException:
        return
    
    content = tweet.find_element_by_xpath('.//div[2]/div[2]/div[1]').text
    likes = tweet.find_element_by_xpath('.//div[@data-testid="like"]').text
    tweet = (users, Date, content, likes)
    return tweet




options = webdriver.ChromeOptions()  ##choose the chrome web browser
options.binary_location = "/Applications/Chrome.app/Contents/MacOS/Google Chrome"  ##The binary location of Chrome
chrome_driver_binary = "/Users/charlesxiong/Downloads/chromedriver-2"  ## The location of chromedriver: an open source tool for automated testing of webapps across many browsers
driver = webdriver.Chrome(chrome_driver_binary, chrome_options=options)  ##Open the driver

driver.get('http://www.twitter.com/login')  #go to log-in page
sleep(2)
username = driver.find_element_by_xpath('//input[@name="session[username_or_email]"]')  #Get the nodes related to username
user_input = input("Enter username: ")
username.send_keys(str(user_input))  #input the username

##Now for the password
askfor_input = getpass()  ##Don't let other see my pw
password = driver.find_element_by_xpath('//input[@name="session[password]"]')
password.send_keys(askfor_input)  ##input pw that I just entered
password.send_keys(Keys.RETURN)  ##click on login botton
sleep(10)

## Let search for some keywords:
search_input = driver.find_element_by_xpath('//input[@aria-label = "Search query"]')  ##get the place to input keywords
keyword = input("Enter your keyword: ")
search_input.send_keys(str(keyword))  ##search for the keyword GameStop
search_input.send_keys(Keys.RETURN)  #click for searching


## Here we are tring to find tweets in a timeslot, so we need to click the "Latest"
driver.find_element_by_link_text('Latest').click()
sleep(5)

#store all data into a list Data
Data = []

#for scrolling
tweet_join = set()

#here we try to get all the tweets info by scrolling down the web page
## We don't know where is the end of the page, so we set a 10 times loop
for scrolling in range(Times_of_scrolling):
    #We first find all sets of nodes related to tweet in that page
    tweets = driver.find_elements_by_xpath('//div[@data-testid="tweet"]')
    
    #Each time we scoll down the web page a little bit.
    ##make sure that we are covering all tweets when scolling down, hence we check the last 20 nodes before this page
    for tweet_node in tweets[-20:]:
        tweet = tweet_info(tweet_node)
        if tweet:
            tweet_item = ', '.join(tweet) ##We create a joined string to combine all the info
            ##if the info doesn't exist in the string, we add them in, so next time we don't duplicate them
            if tweet_item not in tweet_join:
                tweet_join.add(tweet_item)
                Data.append(tweet)
    
    #Now we need to scroll down the page
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
    sleep(2)  #make sure pages are loaded
    
```
